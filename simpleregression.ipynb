### Hello everyone today i have started learning machine learning. In this notebook we will be learning simple regression model.

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
print("imported")
```

Let us now load our file into the program

```python
df = pd.read_csv("homeprices.csv")
df
```

Let us now plot the scatterplot between this two variables basically this is called simple regression technique where we will be using only two variables to train a machine learning model.

```python
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price,color="red",marker="+")
```

Let us now create a regression model using scikit learn library.

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

So, what we have done basically is fitting the area column first and then all the prices for each area into the regression model.

```python
reg.predict([[5000]])
```

As you can see above we have predicted the price of a house of 5000sqft. How is it done?

We used LINEAR EQUATION, 

y = m * x + b


where,

m = slope « reg.coef_

b = intercept « reg.intercept_

x = the value we want to predict price of

Let us see how this worksout

```python
#Let us find the slope first
m = reg.coef_
m
```

```python
#Now intercept
b = reg.intercept_
b
```

```python
#The area we want to predict the size of is 5000 sqft
y = m * 5000 + b
y
```

Let us work on other data set which has only details of areas

```python
df1 = pd.read_csv("areas.csv")
df1
```

```python
price_predict = reg.predict(df1)
price_predict
```

```python
df1["Price"] = price_predict
df1
```

```python
df1.to_csv("prediction.csv")
```
